1- Definición de mi_funcion_sen y mi_funcion_sawtooth

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sp


def mi_funcion_sen( vmax , dc , ff , ph, nn , fs ):
     
    tt = np.linspace( start= 0, stop= (nn-1)/fs ,num = nn)
    xx = vmax*np.sin(tt*2*np.pi*ff+ph) + dc
    return tt, xx

def mi_funcion_sawtooth( vmax , dc , ff , ph, nn , fs ):
     
    tt = np.linspace( start= 0, stop= (nn-1)/fs ,num = nn)
    xx = vmax*sp.sawtooth(tt*2*np.pi*ff+ph) + dc
    return tt, xx
```

2- Ejemplo de implementación de la función sen:

```python
# test función seno
tt, xx = mi_funcion_sen( vmax = 1 , dc = 0 , ff = 1 , ph = 0, nn = 1000 , fs=100 )
fig, ax = plt.subplots()
ax.plot(tt, xx)
ax.set(xlabel='time (s)', ylabel='voltage (V)',title='sen function')
ax.grid()
plt.show()
```

3- Ejemplo de implementación de la función sawtooth:

```python
# test función sawtooth
tt, xx = mi_funcion_sawtooth( vmax = 1 , dc = 0 , ff = 1 , ph = 0, nn = 1000 , fs=100 )
fig, ax = plt.subplots()
ax.plot(tt, xx)
ax.set(xlabel='time (s)', ylabel='voltage (V)',title='sawtooth function')
ax.grid()
plt.show()
```
